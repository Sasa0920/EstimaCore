```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
data.columns
```

```python
data.head(2)
```

```python
features = ["study_hours_per_day","attendance_percentage","mental_health_rating","sleep_hours","part_time_job"]
```

```python
target = "exam_score"
```

```python
data_model=data[features+[target]].copy()
```

```python
data_model
```

```python
le = LabelEncoder()
```

```python
data_model["part_time_job"] = le.fit_transform(data_model["part_time_job"])
```

```python
data_model["part_time_job"].value_counts()
```

```python
X = data_model[features]
y = data_model[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
len(X_train), len(X_test)
```

```python
len(y_train), len(y_test)
```

```python
models = {
  "Linear Regression": {
    "model": LinearRegression(),
    "params": {}
  },
  "Decision Tree": {
    "model": DecisionTreeRegressor(),
    "params": {
      "max_depth": [3,5,10],
      "min_samples_split": [2, 5]
    }
  },
  "Random Forest": {
    "model": RandomForestRegressor(),
    "params": {
      "n_estimators": [50, 100],
      "max_depth": [3, 5],
      "min_samples_split": [2, 5]
    }
  }
}
```

```python
best_models = []
```

```python
for name,config in models.items():
  print(f"Training {name}...")

  #Hyperparameter tuning with GridSearchCV
  grid = GridSearchCV(config["model"], config["params"], cv=5, scoring="neg_mean_squared_error")

  #Fit the model
  grid.fit(X_train, y_train)

  #Evaluate the model
  y_pred = grid.predict(X_test)
  rmse = np.sqrt(mean_squared_error(y_test, y_pred))
  r2 = r2_score(y_test, y_pred)

  best_models.append({
    "model": name,
    "best_params": grid.best_params_,
    "rmse": rmse,
    "r2": r2
  })
  
```

```python
best_models  
```

```python
results = pd.DataFrame(best_models)
```

```python
results
```

```python
results.sort_values(by="rmse")
```

```python
import joblib

best_row = results.sort_values(by="rmse").iloc[0]
```

```python
best_row
```

```python
best_model_name = best_row["model"]
```

```python
best_model_name
```

```python
best_model_config = models[best_model_name]
```

```python
best_model_config
```

```python
final_model = best_model_config["model"].set_params(**best_row["best_params"])
```

```python
final_model.fit(X,y)
```

```python
final_model.predict(X_test)
```

```python
joblib.dump(final_model, "best_model.pkl")
```
